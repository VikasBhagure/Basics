## Set Font property to title and labels

```python
x=np.array([2000,2002,2004,2006,2010])
y=np.array([240,250,260,270,280])

font1={'family':'arial','color':'blue','size':20}
font2={'family':'times new roman','color':'red','size':15}

plt.title("sales data",fontdict=font1)
plt.xlabel("years",fontdict=font2)
plt.ylabel("sales progress",fontdict=font2)
plt.plot(x,y)
plt.show()
```

```python
x=np.array([2002,2004,2006,2008])
y=np.array([240,250,260,280])
plt.title('sales data')
plt.xlabel('years')
plt.ylabel('sales progress')
plt.plot(x,y)
plt.show()
```

## position of title location

```python
x=np.array([2000,2002,2004,2006,2010])
y=np.array([240,250,260,270,280])

font1={'family':'arial','color':'blue','size':20}
font2={'family':'times new roman','color':'red','size':15}

plt.title("sales data",fontdict=font1,loc='right')
plt.xlabel("years",fontdict=font2)
plt.ylabel("sales progress",fontdict=font2)
plt.plot(x,y)
plt.show()
```

## Add Grid Lines To a Plot

```python
from matplotlib import style
style.use ('ggplot')
x=[40,60,82]
y=[12,56,70]
x1=[20,40,50]
y1=[23,34,56]
plt.plot(x,y,'g',label='2019',linewidth=5)
plt.plot(x1,y1,'c',label='2020',linewidth=5)
plt.title('Chart')
plt.ylabel('range')
plt.xlabel('order')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

## Stacked Vrticle Bar

```python
plt.style.use('seaborn-darkgrid')
x=['airtel','jio','vodaphone','idea']
y=[17,18,29,40]
y1=[20,21,22,23]
plt.figure(figsize=(5,7))
plt.bar(x,y,label='2019',width=0.5,color='r')
plt.bar(x,y1,label='2020',width=0.5,bottom=y,color='g')
plt.xlabel('ISP')
plt.ylabel("sales")
plt.title('Sales by  years')
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
x=['airtel','jio','vodaphone','idea']
y=np.array([17,18,29,40])
y1=np.array([20,21,22,23])
y2=np.array([4,5,6,25])
plt.figure(figsize=(5,7))
plt.bar(x,y,label='2019',width=0.5,color='r')
plt.bar(x,y1,label='2020',width=0.5,bottom=y,color='g')
plt.bar(x,y2,label='2021',width=0.5,bottom=y+y1,color='y')
plt.xlabel('ISP')
plt.ylabel("sales")
plt.title('Sales by  years')
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
x=['airtel','jio','vodaphone','idea']
y=[17,18,29,40]
y1=[20,21,22,23]

plt.figure(figsize=(5,7))

plt.barh(x,y,label='2020',color='g')
plt.barh(x,y1,label='2021',left=y,color='y')
plt.xlabel('ISP')
plt.ylabel("sales")
plt.title('Sales by  years')
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
x=['airtel','jio','vodaphone','idea']
y=np.array([17,18,29,40])
y1=np.array([20,21,22,23])
y2=np.array([4,5,6,25])
plt.figure(figsize=(5,7))
plt.barh(x,y,label='2019',color='r')
plt.barh(x,y1,label='2020',left=y,color='g')
plt.barh(x,y2,label='2021',left=y+y1,color='y')
plt.xlabel('ISP')
plt.ylabel("sales")
plt.title('Sales by  years')
plt.legend()
plt.show()
```

## Histogram

```python
age=[6,22,78,13,79,33,8,24,84,90,31,15,52,66,73,21,68,44,65]
bins=[0,20,40,60,80,100]
plt.hist(age,bins,histtype='stepfilled',rwidth=0.6) #rectangular width
plt.xlabel('range')
plt.ylabel('number of people')
plt.title('histogram')
plt.show()
```

## Scatter Plot

```python
x=[1.1,2.2,3.5,4.5,5.2,6.2]
y=[5.7,8.3,9.2,10.1,2.6,3.8]
plt.scatter(x,y,label='high income low salary', color='b',s=200,alpha=0.5)
plt.xlabel('saving')
plt.ylabel('income')
plt.title("scatter plot")

plt.legend()
plt.show()
```

```python
x=[1.1,2.2,3.5,4.5,5.2,6.2]
y=[5.7,8.3,9.2,10.1,2.6,3.8]
size=np.array([200,300,100,50,60,40])
colors=np.array(['red','blue','green','orange','pink','black'])
plt.scatter(x,y,label='high income low salary',c=colors,s=size,alpha=0.5) #alpha used for transperency of scatter dot
plt.xlabel('saving')
plt.ylabel('income')
plt.title("scatter plot")

plt.legend()
plt.show()
```

## Pie Chart

```python
y=np.array([35,25,10,15])
plt.pie(y)
plt.show()
```

```python
y=np.array([35,25,25,15])
l=['airtel','jio','vodaphone','bsnl']
c=['r','b','g','k']
plt.pie(y,labels=l,startangle=90,shadow=True,colors=c,explode=(0,0,0.2,0))
plt.legend(loc='lower left',title='ISP')
plt.show()
```
